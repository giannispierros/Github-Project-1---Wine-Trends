# Cleaning (white)

```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
white_df = pd.read_csv('White.csv', low_memory = False)
white_df.head()
```

```python
white_df.info()
```

```python
white_df.sort_values('Year')
```

```python
white_df['Year'] = white_df['Year'].apply(pd.to_numeric, errors = 'coerce')
white_df.sort_values('Year', ascending = False)
```

```python
white_df.info()
```

```python
white_df = white_df.dropna()
white_df.info()
```

# Analysis (white)

```python
#Top 3 Greek

white_df_greek = white_df[(white_df['Country'] == 'Greece')]
top_greek_white = white_df_greek.sort_values('Rating', ascending = False).head(3)
top_greek_white
```

```python
# White sorted

white_df.sort_values('Price', ascending = False)
```

```python
white_df_international = white_df[(white_df['Price']<=300)]

white_df_international.sort_values('Rating', ascending = False)
```

```python
top_international_white= white_df_international[(white_df_international['NumberOfRatings']>=300)]
```

```python
# Top 10 rated countries for white

top_international_white_avg = top_international_white.groupby('Country').mean()['Rating'].reset_index()
top_international_white_avg.sort_values('Rating', ascending = False).head(5).plot(x = 'Country', y = 'Rating', kind = 'bar', figsize=(12,6))
```

```python
# Cheap white wines with top ratings

cheap_top_international_white = top_international_white[(top_international_white['Price']<=50)].sort_values('Rating', ascending = False).head(5)
cheap_top_international_white
```

```python
# Mid white wines with top ratings

mid_top_international_white = top_international_white[(top_international_white['Price']>50) & (top_international_white['Price']<=150)].sort_values('Rating', ascending = False).head(5)
mid_top_international_white
```

```python
# Expensive white wines with top ratings

exp_top_international_white = top_international_white[(top_international_white['Price']>150)].sort_values('Rating', ascending = False).head(1)
exp_top_international_white
```
